# Estructuración de los datos de Ventanas de tiempo

Agrego 1 columnas a todos los conjuntos de datos llamada 'cryptocurrency', la cual me permite saber cual es la criptomoneda del registro si tiene.

```python
benignoAlexa['Cryptocurrency'] = 'Does_not_have'
bitcoin['Cryptocurrency'] = 'Bitcoin'
bitcash['Cryptocurrency'] = 'Bitcash'
litecoin['Cryptocurrency'] = 'Litecoin'
eth['Cryptocurrency'] = 'Etherium'
monero['Cryptocurrency'] = 'Monero'
sinclasificar['Cryptocurrency'] = 'Unclassified'
```

## Concatenación de datos mineros identificados y datos benignos

```python
pure_data_miner_time = pd.concat([bitcoin,bitcash,litecoin,eth,monero,benignoAlexa], axis=0)
pure_data_miner_time.head()
```

```python
print(colored(f'Reseteo los indices del dataframe, para que no apunte un mismo indice a varios registros', attrs=['bold']))
pure_data_miner_time = pure_data_miner_time.reset_index()
```

```python
print(colored(f'Genero un rango de valores', attrs=['bold']))
indices = list(pure_data_miner_time.index)
print(colored(f'El rango de valores se le aplico aleatoriedad de manera pseudoaleatoria', attrs=['bold']))
random.shuffle(indices)
```

```python
print(colored(f'Reacomodo los registros del conjunto con los indices aleatorios', attrs=['bold']))
pure_data_miner_time = pure_data_miner_time.reindex(indices)
pure_data_miner_time
```

```python
print(colored(f'Guardo el conjunto de datos', attrs=['bold']))
pure_data_miner_time.to_csv('dataC.csv', index=False)
```
