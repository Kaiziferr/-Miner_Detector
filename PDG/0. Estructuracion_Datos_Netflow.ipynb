# Estructuración de los datos de Netflow

Agrego 2 columnas a todos los conjuntos de datos. Primero 'cryptocurrency' me permite saber cual es la criptomoneda del registro si tiene. La ultima para saber si el registro corresponde a un minero o no

```python
benignoAlexa_f['cryptocurrency'] = 'Does_not_have'
benignoAlexa_f['type'] = 'not_mine'
bitcoin_f['cryptocurrency'] = 'Bitcoin'
bitcoin_f['type'] = 'mine'
bitcash_f['cryptocurrency'] = 'Bitcash'
bitcash_f['type'] = 'mine'
litecoin_f['cryptocurrency'] = 'Litecoin'
litecoin_f['type'] = 'mine'
eth_f['cryptocurrency'] = 'Etherium'
eth_f['type'] = 'mine'
monero_f['cryptocurrency'] = 'Monero'
monero_f['type'] = 'mine'
sin_clasificar_f['cryptocurrency'] = 'Unclassified'
sin_clasificar_f['type'] = 'mine'
```

## Concatenación de datos mineros identificados y datos benignos

```python
pure_data_miner_netflow = pd.concat([bitcoin_f,bitcash_f,litecoin_f,eth_f,monero_f,benignoAlexa_f], axis=0)
pure_data_miner_netflow.head()
```

```python
print(colored(f'Reseteo los indices del dataframe, para que no apunte un mismo indice a varios registros', attrs=['bold']))
pure_data_miner_netflow = pure_data_miner_netflow.reset_index()
```

```python
print(colored(f'Genero un rango de valores', attrs=['bold']))
indices = list(pure_data_miner_netflow.index)
print(colored(f'El rango de valores se le aplico aleatoriedad de manera pseudoaleatoria', attrs=['bold']))
random.shuffle(indices)
```

```python
print(colored(f'Reacomodo los registros del conjunto con los indices aleatorios', attrs=['bold']))
pure_data_miner_netflow = pure_data_miner_netflow.reindex(indices)
pure_data_miner_netflow
```

```python
print(colored(f'Guardo el conjunto de datos', attrs=['bold']))
pure_data_miner_netflow.to_csv('dataA.csv',index=False)
```
