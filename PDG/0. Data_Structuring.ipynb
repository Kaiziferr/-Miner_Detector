Asignamos una columna 'Cryptocurrency' y 'Id_Cryptocurrency' para representar el tipo de criptomoneda, ya que concatenaremos los conjunto de datos mineros y el conjunto de trafico no minero.

```python
benignoAlexa['Cryptocurrency'] = None
benignoAlexa['Id_Cryptocurrency'] = 0
```

```python
bitcoin['Cryptocurrency'] = 'Bitcoin'
bitcoin['Id_Cryptocurrency'] = 1
bitcash['Cryptocurrency'] = 'Bitcash'
bitcash['Id_Cryptocurrency'] = 2
litecoin['Cryptocurrency'] = 'Litecoin'
litecoin['Id_Cryptocurrency'] = 3
eth['Cryptocurrency'] = 'Etherium'
eth['Id_Cryptocurrency'] = 4
monero['Cryptocurrency'] = 'Monero'
monero['Id_Cryptocurrency'] = 5
sinclasificar['Cryptocurrency'] = 'sinclasificar'
sinclasificar['Id_Cryptocurrency'] = 6
```

#### Concatenaremos los conjuntos de datos de trafico minero y del conjunto no aleatorio de trafico no minero

```python
pure_data_miner = pd.concat([bitcoin,bitcash,litecoin,eth,monero,benignoAlexa], axis=0)
```

```python
pure_data_miner.index
```

```python
print(f"El dataset estructurado inicial es de {pure_data_miner.ndim} dimensiones con {pure_data_miner.shape[0]} filas y {pure_data_miner.shape[1]} columnas")
```

```python
pure_data_miner.head()
```

```python
# Resetie los indices del dataframe, para que no apunte un mismo indice a varios registros, y hacemos una copia para no afectar la integridad de la copia original
pure_data_miner = pure_data_miner.reset_index()
```

```python
# Genere un rango de valores
indices = list(range(0,len(pure_data_miner.index)))
import random
# El rango de valores se le aplico aleatoriedad, la desorganizar de manera pseudoaleatorio los elementos
random.shuffle(indices)
```

```python
# Introduje aleatoriedad al conjunto de datos con el rango de indices 
pure_data_miner = pure_data_miner.reindex(indices)
pure_data_miner
```
